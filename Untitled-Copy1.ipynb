```python
import math

# 第 1 步你手动拿到的 4 个真实数字
N        = 20328575   # 数据库总报告数
a_plus_b = 76216      # 所有顺铂报告
a_plus_c = 36540      # 所有耳聋报告
a        = 624        # 顺铂 且 耳聋

# 反推 2x2 表其余三格
b = a_plus_b - a
c = a_plus_c - a
d = N - a - b - c

# 计算 ROR 和 95% 置信区间
ror = (a * d) / (b * c)
se  = math.sqrt(1/a + 1/b + 1/c + 1/d)
ror_low  = math.exp(math.log(ror) - 1.96 * se)
ror_high = math.exp(math.log(ror) + 1.96 * se)

print(f"ROR = {ror:.2f}   95%CI ({ror_low:.2f}, {ror_high:.2f})")
print("信号成立" if ror_low > 1 and a >= 3 else "无信号")
```

```python
import requests

BASE = "https://api.fda.gov/drug/event.json"

def get_total(search=None):
    """给一个查询条件，返回 openFDA 里匹配的报告总数。不给条件就返回数据库总数。"""
    params = {"limit": 1}
    if search:
        params["search"] = search
    r = requests.get(BASE, params=params, timeout=30)
    if r.status_code == 404:      # openFDA 查不到任何记录时会返回 404
        return 0
    r.raise_for_status()          # 其他错误（如限流）就抛出来，方便发现问题
    return r.json()["meta"]["results"]["total"]
```

```python
import math

# 定义"药"和"事件"的查询字段（就是你第 1 步网址里 search= 后面那部分）
drug  = 'patient.drug.openfda.generic_name:"cisplatin"'
event = 'patient.reaction.reactionmeddrapt:"deafness"'

# 让代码自己去 openFDA 取回 4 个数字
N        = get_total()                          # 数据库总报告数
a_plus_b = get_total(drug)                      # 所有顺铂报告
a_plus_c = get_total(event)                     # 所有耳聋报告
a        = get_total(f"{drug} AND {event}")     # 顺铂 且 耳聋

print(f"自动取回：N={N}, a+b={a_plus_b}, a+c={a_plus_c}, a={a}")

# 下面这段和第 2 步完全一样
b = a_plus_b - a
c = a_plus_c - a
d = N - a - b - c
ror = (a * d) / (b * c)
se  = math.sqrt(1/a + 1/b + 1/c + 1/d)
ror_low  = math.exp(math.log(ror) - 1.96 * se)
ror_high = math.exp(math.log(ror) + 1.96 * se)

print(f"ROR = {ror:.2f}   95%CI ({ror_low:.2f}, {ror_high:.2f})")
print("信号成立" if ror_low > 1 and a >= 3 else "无信号")
```

```python
import math

def analyze(drug, event):
    """给一个药名 + 一个事件词，返回这一格的 ROR / 95%CI / 是否成信号。"""
    N        = get_total("*")   # 总报告数（* 匹配全部）
    a_plus_b = get_total(f'patient.drug.openfda.generic_name:"{drug}"')
    a_plus_c = get_total(f'patient.reaction.reactionmeddrapt:"{event}"')
    a = get_total(f'patient.drug.openfda.generic_name:"{drug}"'
                  f' AND patient.reaction.reactionmeddrapt:"{event}"')
    b = a_plus_b - a
    c = a_plus_c - a
    d = N - a - b - c

    ror = (a * d) / (b * c)
    se  = math.sqrt(1/a + 1/b + 1/c + 1/d)
    low  = math.exp(math.log(ror) - 1.96 * se)
    high = math.exp(math.log(ror) + 1.96 * se)
    signal = low > 1 and a >= 3          # 信号判定：CI 下限 > 1 且 a ≥ 3

    return {"drug": drug, "event": event, "a": a,
            "ROR": round(ror, 2), "CI_low": round(low, 2),
            "CI_high": round(high, 2), "signal": signal}

# 拿已知答案验一下：顺铂 × 耳聋，应得 ROR≈4.65
print(analyze("cisplatin", "deafness"))
```

```python
import requests, math, time

API_KEY = None      # 有 openFDA key 就填进引号里，没有就保持 None
SLEEP = 0.3         # 每次查询后歇 0.3 秒，别把 FDA 服务器问烦了
BASE = "https://api.fda.gov/drug/event.json"
print("工具准备好了")
```

```python

def _count(search_query, retries=3):
    url = f"{BASE}?search={search_query}&limit=1"
    if API_KEY: url += f"&api_key={API_KEY}"
    for attempt in range(retries):
        try:
            r = requests.get(url, timeout=30)      # 最多等 30 秒，别无限挂着
            time.sleep(SLEEP)
            if r.status_code == 404:               # 没匹配到，当作 0
                return 0
            if r.status_code == 429:               # 被限流，多等一会再试
                time.sleep(2 * (attempt + 1))
                continue
            r.raise_for_status()
            return r.json()["meta"]["results"]["total"]
        except requests.exceptions.RequestException:
            if attempt == retries - 1:             # 试满 3 次还不行，才真的报错
                raise
            time.sleep(2 * (attempt + 1))          # 等 2 秒、4 秒…再试

def _total_db():
    url = f"{BASE}?limit=1"
    if API_KEY: url += f"&api_key={API_KEY}"
    r = requests.get(url); time.sleep(SLEEP); r.raise_for_status()
    return r.json()["meta"]["results"]["total"]

def reaction_or_clause(pt_list):
    # 把一组 PT 拼成 OR；.exact 保证整词匹配，词里的空格换成 +
    terms = "+".join(f'"{pt.replace(" ", "+")}"' for pt in pt_list)
    return f"patient.reaction.reactionmeddrapt.exact:({terms})"

def drug_clause(name):
    # ↓↓↓ 这一行必须和你旧代码用的药物字段一致；对不上 4.65 就改这里 ↓↓↓
    return f'patient.drug.openfda.generic_name.exact:"{name}"'
    # 备选（若你旧代码用自由文本）：
    # return f'patient.drug.medicinalproduct.exact:"{name}"'

def ror_prr(drug, pt_list, N=None):
    d = drug_clause(drug); e = reaction_or_clause(pt_list)
    a = _count(f"{d}+AND+{e}")          # 药 且 事件
    drug_total = _count(d)              # 药（a+b）
    event_total = _count(e)             # 事件（a+c）
    if N is None: N = _total_db()       # 总报告数
    b = drug_total - a; c = event_total - a; dd = N - drug_total - event_total + a
    if a < 3 or b<=0 or c<=0 or dd<=0:
        return {"药":drug,"a":a,"b":b,"c":c,"d":dd,"说明":"样本不足或空单元"}
    ror = (a*dd)/(b*c)
    se = math.sqrt(1/a+1/b+1/c+1/dd)
    lo = math.exp(math.log(ror)-1.96*se); hi = math.exp(math.log(ror)+1.96*se)
    prr = (a/(a+b))/(c/(c+dd))
    return {"药":drug,"PT数":len(pt_list),"a":a,"ROR":round(ror,2),
            "95%CI":(round(lo,2),round(hi,2)),"PRR":round(prr,2),"信号":(lo>1 and a>=3)}

print("引擎装好了")
```

```python
OTOTOXICITY = {
    "cochlear_core": [   # 耳蜗/听觉（主分析用）
        "DEAFNESS","DEAFNESS NEUROSENSORY","DEAFNESS BILATERAL",
        "DEAFNESS UNILATERAL","DEAFNESS PERMANENT","SUDDEN HEARING LOSS",
        "HYPOACUSIS","TINNITUS","OTOTOXICITY",
    ],
    "vestibular_core": ["VERTIGO","VERTIGO POSITIONAL","BALANCE DISORDER"],
    "cochlear_extended": ["HYPERACUSIS","AUDITORY DISTURBANCE"],   # 只做敏感性分析
    "vestibular_extended": ["DIZZINESS","LABYRINTHITIS"],          # 只做敏感性分析，就是主分析只用 core 里的词，但怕漏掉信号，所以额外再扩展几个相关词跑一遍看看结果稳不稳定。打个比方：你查"头痛"（core），再扩到"头晕"（extended）看看结果是不是差不多。
    "anchor_pt": "DEAFNESS",   # 复现旧结果的锚点
    
    # ——1b 最终定论：耳毒性主口径 = 特异听力核心（不拆碎、不稀释）——
    "primary_cochlear": [
        "DEAFNESS","DEAFNESS NEUROSENSORY","NEUROSENSORY HYPOACUSIS",
        "DEAFNESS BILATERAL","DEAFNESS UNILATERAL","DEAFNESS PERMANENT",
        "DEAFNESS TRANSITORY","SUDDEN HEARING LOSS","HYPOACUSIS","OTOTOXICITY",
    ],
    # 前庭作为独立副维度单独报告，不并入主口径
    "primary_vestibular": ["VERTIGO","VERTIGO POSITIONAL","BALANCE DISORDER"],
}
MEDDRA_VERSION = "SMQ 20000171 (Hearing impairment) narrow子集; 依 Liu et al. 2025, Front Pharmacol, Table S1 核对"
# 说明：主口径以官方 SMQ 20000171 为全集，取感觉神经性药物耳毒子集，剔除 presbyacusis 等混杂/结构/检查类 PT   # 用 MedDRA 核对术语后填版本号   #MedDRA 是一个标准医学词典，不良反应术语是从这本词典里取的。不同版本可能词不完全一样，所以记下用的是哪个版本，方便别人复现你的结果。现在先用 TODO 占个位。

def pt_set(tox, include_extended=False, dimension="all"):   #（9 个词，含 TINNITUS）——这是你做探索、做对比时用的。还记得你跑那个 ①–⑤ 五个口径的对比吗？其中 ② 号"仅耳蜗 core"、④ 号"全部 core"，用的就是这套。它含 tinnitus、含前庭，是故意留着"宽一点"的版本，好让你看出"加了不特异的词会不会稀释信号"。它的价值在"对比"，不在"最终用"。
    pts = []
    if dimension in ("all","cochlear"):
        pts += tox["cochlear_core"] + (tox["cochlear_extended"] if include_extended else [])
    if dimension in ("all","vestibular"):
        pts += tox["vestibular_core"] + (tox["vestibular_extended"] if include_extended else [])
    return pts

print("core 一共", len(pt_set(OTOTOXICITY)), "个 PT：", pt_set(OTOTOXICITY))
# 最终定稿口径，正式分析只用这个变量
FINAL_COCHLEAR = OTOTOXICITY["primary_cochlear"]
```
