```python
"""
signal_pipeline.py
==================
信号引擎（不变的部分）。它不关心是什么毒性，只接收一组 PT，算 ROR/PRR。
换毒性 = 换传进来的 PT 列表；这个文件永远不用改。

1b 相对旧代码只改了一处：事件那一句从"单个 PT"变成"一组 PT 的 OR"。
药物那句、ROR 公式、openFDA 调用都不变。

【未测试声明】我这边的网络访问不到 api.fda.gov，无法替你实跑。
请用你的"已知答案"校验：
  第 1 步：pt_list=["DEAFNESS"] 跑顺铂，应复现你旧的 ROR≈4.65（证明代码没错）
  第 2 步：换成整组 SMQ 的 PT，再看结果（会不同、但应仍强阳性）
只有第 1 步对上了，第 2 步的数才可信。这是你一贯的纪律。
"""

import math
import time
import requests

BASE = "https://api.fda.gov/drug/event.json"
# 如果申请了 openFDA API key，填在这里可提高配额（否则留 None）
API_KEY = None
SLEEP = 0.3   # 礼貌限速，避免超过 240 请求/分钟


def _count(search_query):
    """
    返回匹配某 search 条件的报告数（meta.results.total）。

    关键：手动拼 URL 字符串，不用 requests 的 params=，
    否则 '+AND+' 会被二次编码成 '%2BAND%2B'，静默返回 0 —— 你之前踩过的坑。
    openFDA 无匹配时返回 HTTP 404，这里当作 0 处理。
    """
    url = f"{BASE}?search={search_query}&limit=1"
    if API_KEY:
        url += f"&api_key={API_KEY}"
    r = requests.get(url)
    time.sleep(SLEEP)
    if r.status_code == 404:
        return 0
    r.raise_for_status()
    return r.json()["meta"]["results"]["total"]


def _total_db():
    """数据库总报告数 N（不带 search）。"""
    url = f"{BASE}?limit=1"
    if API_KEY:
        url += f"&api_key={API_KEY}"
    r = requests.get(url)
    time.sleep(SLEEP)
    r.raise_for_status()
    return r.json()["meta"]["results"]["total"]


def reaction_or_clause(pt_list):
    """
    把一组 PT 拼成 openFDA 的 OR 条件（这就是 1b 唯一的新东西）。
    括号内空格(URL 里是 +)分隔 = OR；.exact 保证整词匹配，
    否则 'SUDDEN HEARING LOSS' 会被拆成 SUDDEN / HEARING / LOSS 分别计数。
    """
    quoted = "+".join(f'"{pt}"' for pt in pt_list)
    return f"patient.reaction.reactionmeddrapt.exact:({quoted})"


def drug_clause(generic_name):
    """
    药物条件。这一句请与你【已跑通 4.65 的旧代码】保持一致：
    如果你旧代码用的是 medicinalproduct 自由文本，就改用那个字段，
    否则 N/分母不同会导致 ROR 对不上。默认用规范化的 generic_name。
    """
    return f'patient.drug.openfda.generic_name.exact:"{generic_name}"'
    # 备选（若旧代码用自由文本）：
    # return f'patient.drug.medicinalproduct.exact:"{generic_name}"'


def ror_prr(drug_generic, pt_list, N=None):
    """
    对 (药 × 一组事件PT) 算 ROR、PRR 及 95%CI，返回 2x2 与结果。
    信号判定沿用你的规则：CI 下限 > 1 且 a >= 3。
    """
    d = drug_clause(drug_generic)
    e = reaction_or_clause(pt_list)

    a = _count(f"{d}+AND+{e}")          # 药 且 事件
    drug_total = _count(d)              # 药 (=a+b)
    event_total = _count(e)             # 事件 (=a+c)
    if N is None:
        N = _total_db()                 # 总报告数

    b = drug_total - a
    c = event_total - a
    dcell = N - drug_total - event_total + a

    if a < 3 or b <= 0 or c <= 0 or dcell <= 0:
        return {"drug": drug_generic, "a": a, "b": b, "c": c, "d": dcell,
                "note": "样本不足或空单元，信号不成立(需 a>=3)"}

    ror = (a * dcell) / (b * c)
    se = math.sqrt(1/a + 1/b + 1/c + 1/dcell)
    ror_low = math.exp(math.log(ror) - 1.96 * se)
    ror_high = math.exp(math.log(ror) + 1.96 * se)

    prr = (a / (a + b)) / (c / (c + dcell))

    return {
        "drug": drug_generic,
        "n_pts": len(pt_list),
        "a": a, "b": b, "c": c, "d": dcell,
        "ROR": round(ror, 2),
        "ROR_95CI": (round(ror_low, 2), round(ror_high, 2)),
        "PRR": round(prr, 2),
        "signal": (ror_low > 1 and a >= 3),
    }


if __name__ == "__main__":
    from toxicity_definitions import OTOTOXICITY, pt_set

    # ---------- 第 1 步：已知答案校验（必须先过）----------
    # 用单个 DEAFNESS 复现你旧的 ROR≈4.65。对上 → 代码可信。
    print(">> 校验：顺铂 × 单个 DEAFNESS（应≈你的 4.65）")
    print(ror_prr("CISPLATIN", [OTOTOXICITY["anchor_pt"]]))

    # ---------- 第 2 步：升级到 SMQ 整组 ----------
    print("\n>> 升级：顺铂 × 耳毒性 SMQ(core, 全维度)")
    print(ror_prr("CISPLATIN", pt_set(OTOTOXICITY, include_extended=False)))

    # ---------- 可选：分维度看（耳蜗 vs 前庭）----------
    print("\n>> 顺铂 × 仅耳蜗")
    print(ror_prr("CISPLATIN", pt_set(OTOTOXICITY, dimension="cochlear")))
    print(">> 顺铂 × 仅前庭")
    print(ror_prr("CISPLATIN", pt_set(OTOTOXICITY, dimension="vestibular")))
```
